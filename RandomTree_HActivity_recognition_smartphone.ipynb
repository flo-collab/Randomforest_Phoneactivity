```python
import pandas as pd
import timeit
train = pd.read_csv('/content/drive/MyDrive/Datasets/Human Activity Recognition with Smartphones/train.csv')
test = pd.read_csv('/content/drive/MyDrive/Datasets/Human Activity Recognition with Smartphones/test.csv')
```

```python
print(train.info())
print(train.describe())
print('La taille du dataset train est:',train.shape)
train.head()
```

```python
#On sépare notre jeu de données
X_train = train[train.columns[:-2]]
y_train = train['Activity']

X_test = test[test.columns[:-2]]
y_test = test['Activity']
```

```python
#Selection du modele
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 500, oob_score = True)
start_time = timeit.default_timer()
#Entrainement du Modele sur les données brutes
model = rfc.fit(X_train, y_train)
elapsed1 = timeit.default_timer() - start_time
```

```python
from sklearn.metrics import accuracy_score
# On regarde les performances de notre modele 
pred = rfc.predict(X_test)
accuracy1 = accuracy_score(y_test, pred)
print("accuracy {:.2f} time {:.2f}s".format(accuracy1, elapsed1))
```

```python
from sklearn.feature_selection import SelectFromModel
# On selectionne les features les plus importantes  
select = SelectFromModel(rfc, prefit=True, threshold=0.003)
X_train2 = select.transform(X_train)
print(X_train2.shape)
```

On est passé de 562 features à 84 features.
Comparons maintenant les temps de calcul

```python
rfc2 = RandomForestClassifier(n_estimators=500, oob_score=True)

start_time = timeit.default_timer()

rfc2 = rfc2.fit(X_train2, y_train)

X_test2 = select.transform(X_test)

elapsed2 = timeit.default_timer() - start_time
pred = rfc2.predict(X_test2)

accuracy2 = accuracy_score(y_test, pred)
print("accuracy avec toutes les features : {:.2f} time {:.2f}s".format(accuracy1, elapsed1))
print("accuracy avec nombre de features reduit {:.2f} time {:.2f}s".format(accuracy2, elapsed2))
```
